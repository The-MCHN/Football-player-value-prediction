```python
import pandas as pd
```

```python
df = pd.read_csv('BestFootballersDataSet.csv')
```

```python
df.head()
```

```python
def fix_position_col(x):
    return x.split('"')[0]
```

```python
df['position'] = df['position'].apply(lambda x : fix_position_col(x))
```

```python
df.head()
```

```python
type(df.value_euro)
```

```python
def fix_value_col(x):
    if 'k' in str(x):
        return int(str(x).split('k')[0]+ '000')
    return int(str(x).split('.')[0])
```

```python
df['value_euro'] = df['value_euro'].apply(lambda x : fix_value_col(x))
```

```python
df['value_euro']
```

```python
df.to_csv("./BestFootballersDataSet.csv", index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(df, x='value_euro', bins=30)
plt.show()
```
